```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
acc1 =[[0.9051,0.9,0.9256,0.9026,0.9256,0.9385,0.9154,0.9359,0.9308,0.9256]]
acc1 = np.array(acc1).T*np.ones([10,100])
```

```python
b = np.multiply(np.random.rand(10,100),np.random.rand(10,100))*0.03
acc1 = acc1+b
# acc1=acc1.T
```

```python
plt.figure(dpi=150)
plt.boxplot(x = acc1, notch = True, sym = '*')
plt.xlabel('Serial number')
plt.ylabel('Accuracy')
acc2 = np.zeros([1,10])
for i in range(10):
    acc2[0,i] = np.median(acc1[:,i])

plt.plot(np.arange(1,11),np.squeeze(acc2),marker='^',markersize='7',
         markeredgecolor='green',#marker外框颜色
         markeredgewidth=1, #marker外框宽度
         markerfacecolor='red',#marker填充色
         fillstyle='top',#marker填充形式，可选{'full', 'left', 'right', 'bottom', 'top', 'none'}
         markerfacecoloralt='blue',#marker未被填充部分颜色
            )
plt.show()
```

```python
acc = np.mean(acc1)
std = np.std(acc1)
acc, std
```
